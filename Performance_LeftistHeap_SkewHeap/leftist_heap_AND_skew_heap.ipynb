```python
# common way to handle if file does not exist
try:
    f = open("result.txt")
except IOError:
    print('The file result.txt does not exist')
    
LH = []
SH = []
for line in f:
    if(line[0] == 'L'):
        LH.append(line)
    elif(line[0] == 'S'):
        SH.append(line)

for i in LH:
    print(i)
for j in SH:
    print(j)
```

```python
print(len(LH))
print(len(SH))
```

```python
LH_build = {}
LH_build[50000] = []
LH_build[100000] = []
LH_build[200000] = []
LH_build[400000] = []

LH_operation = {}
LH_operation[50000] = []
LH_operation[100000] = []
LH_operation[200000] = []
LH_operation[400000] = []

for i, j in enumerate(LH):
    seconds_build = j.rstrip()[32:39]
    seconds_operation = j.rstrip()[36:44]
    if(i+1 <= 10): # n = 50000
        if "build" in LH[i]:
            LH_build[50000].append(seconds_build)
        else:
            LH_operation[50000].append(seconds_operation)
    elif(i+1 >10 and i+1 <= 20):
        if "build" in LH[i]:
            LH_build[100000].append(seconds_build)
        else:
            LH_operation[100000].append(seconds_operation)
    elif(i+1 >20 and i+1 <= 30):
        if "build" in LH[i]:
            LH_build[200000].append(seconds_build)
        else:
            LH_operation[200000].append(seconds_operation)
    else:
        if "build" in LH[i]:
            LH_build[400000].append(seconds_build)
        else:
            LH_operation[400000].append(seconds_operation)
            
print(LH_build)
print(LH_operation)
```

```python
SH_build = {}
SH_build[50000] = []
SH_build[100000] = []
SH_build[200000] = []
SH_build[400000] = []

SH_operation = {}
SH_operation[50000] = []
SH_operation[100000] = []
SH_operation[200000] = []
SH_operation[400000] = []

for i, j in enumerate(SH):
    seconds_build = j.rstrip()[29:37]
    seconds_operation = j.rstrip()[33:41]
    if(i+1 <= 10): # n = 50000
        if "build" in SH[i]:
            SH_build[50000].append(seconds_build)
        else:
            SH_operation[50000].append(seconds_operation)
    elif(i+1 >10 and i+1 <= 20): # n = 100,000
        if "build" in SH[i]:
            SH_build[100000].append(seconds_build)
        else:
            SH_operation[100000].append(seconds_operation)
    elif(i+1 >20 and i+1 <= 30):
        if "build" in SH[i]:
            SH_build[200000].append(seconds_build)
        else:
            SH_operation[200000].append(seconds_operation)
    else:
        if "build" in SH[i]:
            SH_build[400000].append(seconds_build)
        else:
            SH_operation[400000].append(seconds_operation)
            
print(SH_build)
print(SH_operation)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
plt.title("Build with n = 50,000")
plt.xlabel("# of trials")
plt.ylabel("Seconds")
plt.xticks(x)
plt.plot(x, LH_build[50000], "ro", x, SH_build[50000], "bs")
plt.axis([0.5, 5.5, 0, 0.06])


avg_b = sum(map(float, LH_build[50000])) / 5
avg_h = sum(map(float, SH_build[50000])) / 5

plt.text(2, 0.05345, "Leftist Heap build average: " + str(avg_b)[0:6] + " seconds")
plt.text(2, 0.0125, "Skew Heap build average: "  + str(avg_h)[0:6] + " seconds")

fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/build50000.png", dpi = 100)

plt.show()
```

```python
def avg(list):
    return sum(map(float, list)) / 5

generated = [50000, 100000, 200000, 400000]

# get the data as a dict with generated associated with its average
data_lhb = {}
data_shb = {}
data_lho = {}
data_sho = {}

for num_gen in generated:
    data_lhb[num_gen] = avg(LH_build[num_gen])
    data_shb[num_gen] = avg(SH_build[num_gen])
    data_lho[num_gen] = avg(LH_operation[num_gen])
    data_sho[num_gen] = avg(SH_operation[num_gen])
    
    
print(data_lhb)
print(data_shb)
print(data_lho)
print(data_sho)
```

```python
#----------plotting begins--------------

plt.title("Comparison of Build between Leftist Heap and Skew Heap")
plt.xlabel("# of random number generations to insert into data structure")
plt.ylabel("Seconds")
plt.xticks(generated)

lists_lhb = sorted(data_lhb.items())
x1, y1 = zip(*lists_lhb) # unpack a list of pairs into two tuples
lists_shb = sorted(data_shb.items())
x2, y2 = zip(*lists_shb)


plt.text(200000, 0.35, "Leftist Heap")
plt.text(200000, 0.065, "Skew Heap")

plt.plot(x1, y1, "r", x2, y2, "b")

fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/build_compare.png", dpi = 100)

plt.show()
```

```python
#----------plotting begins--------------

plt.title("Comparison of Operations between Leftist Heap and Skew Heap")
plt.xlabel("# of random number generations with operations")
plt.ylabel("Seconds")
plt.xticks(generated)

lists_lho = sorted(data_lho.items())
x1, y1 = zip(*lists_lho) # unpack a list of pairs into two tuples
lists_sho = sorted(data_sho.items())
x2, y2 = zip(*lists_sho)

plt.text(200000, 0.157, "Leftist Heap")
plt.text(200000, 0.025, "Skew Heap")

plt.plot(x1, y1, "r", x2, y2, "b")


fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/operation_compare.png", dpi = 100)

plt.show()
```
