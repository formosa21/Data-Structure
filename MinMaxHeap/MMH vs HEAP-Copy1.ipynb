```python
# common way to handle if file does not exist
try:
    f = open("result.txt")
except IOError:
    print('The file result.txt does not exist')
    
MMH = []
Heap = []
for line in f:
    if(line[0] == 'M'):
        MMH.append(line)
    elif(line[0] == 'H'):
        Heap.append(line)

for i in MMH:
    print(i)
for j in Heap:
    print(j)
```

```python
print(len(MMH))
print(len(Heap))
```

```python
MMH_build = {}
MMH_build[50000] = []
MMH_build[100000] = []
MMH_build[200000] = []
MMH_build[400000] = []

MMH_operation = {}
MMH_operation[50000] = []
MMH_operation[100000] = []
MMH_operation[200000] = []
MMH_operation[400000] = []

for i, j in enumerate(MMH):
    seconds_build = j.rstrip()[30:37]
    seconds_operation = j.rstrip()[34:42]
    if(i+1 <= 10): # n = 50000
        if "build" in MMH[i]:
            MMH_build[50000].append(seconds_build)
        else:
            MMH_operation[50000].append(seconds_operation)
    elif(i+1 >10 and i+1 <= 20):
        if "build" in MMH[i]:
            MMH_build[100000].append(seconds_build)
        else:
            MMH_operation[100000].append(seconds_operation)
    elif(i+1 >20 and i+1 <= 30):
        if "build" in MMH[i]:
            MMH_build[200000].append(seconds_build)
        else:
            MMH_operation[200000].append(seconds_operation)
    else:
        if "build" in MMH[i]:
            MMH_build[400000].append(seconds_build)
        else:
            MMH_operation[400000].append(seconds_operation)
            
print(MMH_build)
print(MMH_operation)
```

```python
Heap_build = {}
Heap_build[50000] = []
Heap_build[100000] = []
Heap_build[200000] = []
Heap_build[400000] = []

Heap_operation = {}
Heap_operation[50000] = []
Heap_operation[100000] = []
Heap_operation[200000] = []
Heap_operation[400000] = []

for i, j in enumerate(Heap):
    seconds_build = j.rstrip()[24:31]
    seconds_operation = j.rstrip()[28:35]
    if(i+1 <= 10): # n = 50000
        if "build" in Heap[i]:
            Heap_build[50000].append(seconds_build)
        else:
            Heap_operation[50000].append(seconds_operation)
    elif(i+1 >10 and i+1 <= 20): # n = 100,000
        if "build" in Heap[i]:
            Heap_build[100000].append(seconds_build)
        else:
            Heap_operation[100000].append(seconds_operation)
    elif(i+1 >20 and i+1 <= 30):
        if "build" in Heap[i]:
            Heap_build[200000].append(seconds_build)
        else:
            Heap_operation[200000].append(seconds_operation)
    else:
        if "build" in Heap[i]:
            Heap_build[400000].append(seconds_build)
        else:
            Heap_operation[400000].append(seconds_operation)
            
print(Heap_build)
print(Heap_operation)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
plt.title("Build with n = 50,000")
plt.xlabel("# of trials")
plt.ylabel("Seconds")
plt.xticks(x)
plt.plot(x, MMH_build[50000], "ro", x, Heap_build[50000], "bs")
plt.axis([0.5, 5.5, 0, 0.018])


avg_b = sum(map(float, MMH_build[50000])) / 5
avg_h = sum(map(float, Heap_build[50000])) / 5

plt.text(2.5, 0.00245, "MMH build average: " + str(avg_b)[0:6] + " seconds")
plt.text(2.5, 0.00125, "Heap build average: "  + str(avg_h)[0:6] + " seconds")

fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/build50000.png", dpi = 100)

plt.show()
```

```python
def avg(list):
    return sum(map(float, list)) / 5

generated = [50000, 100000, 200000, 400000]

# get the data as a dict with generated associated with its average
data_Mb = {}
data_Hb = {}
data_Mo = {}
data_Ho = {}

for num_gen in generated:
    data_Mb[num_gen] = avg(MMH_build[num_gen])
    data_Hb[num_gen] = avg(Heap_build[num_gen])
    data_Mo[num_gen] = avg(MMH_operation[num_gen])
    data_Ho[num_gen] = avg(Heap_operation[num_gen])
    
    
print(data_Mb)
print(data_Hb)
print(data_Mo)
print(data_Ho)
```

```python
#----------plotting begins--------------

plt.title("Comparison of Build between MMH and Heap")
plt.xlabel("# of random number generations to insert into data structure")
plt.ylabel("Seconds")
plt.xticks(generated)

lists_Mb = sorted(data_Mb.items())
x1, y1 = zip(*lists_Mb) # unpack a list of pairs into two tuples
lists_Hb = sorted(data_Hb.items())
x2, y2 = zip(*lists_Hb)


plt.text(200000, 0.01, "MMH")
plt.text(200000, 0.0045, "Heap")

plt.plot(x1, y1, "r", x2, y2, "b")

fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/build_compare.png", dpi = 100)

plt.show()
```

```python
#----------plotting begins--------------

plt.title("Comparison of Operations between MMH and Heap")
plt.xlabel("# of random number generations with operations")
plt.ylabel("Seconds")
plt.xticks(generated)

lists_Mo = sorted(data_Mo.items())
x1, y1 = zip(*lists_Mo) # unpack a list of pairs into two tuples
lists_Ho = sorted(data_Ho.items())
x2, y2 = zip(*lists_Ho)

plt.text(200000, 0.003, "MMH")
plt.text(200000, 0.007, "Heap")

plt.plot(x1, y1, "r", x2, y2, "b")


fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/operation_compare.png", dpi = 100)

plt.show()
```
