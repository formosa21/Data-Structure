```python
# common way to handle if file does not exist
try:
    f = open("result.txt")
except IOError:
    print('The file result.txt does not exist')
    
BST = []
Heap = []
for line in f:
    if(line[0] == 'B'):
        BST.append(line)
    elif(line[0] == 'H'):
        Heap.append(line)

for i in BST:
    print(i)
for j in Heap:
    print(j)
```

```python
print(len(BST))
print(len(Heap))
```

```python
BST_build = {}
BST_build[50000] = []
BST_build[100000] = []
BST_build[200000] = []
BST_build[400000] = []

BST_operation = {}
BST_operation[50000] = []
BST_operation[100000] = []
BST_operation[200000] = []
BST_operation[400000] = []

for i, j in enumerate(BST):
    seconds_build = j.rstrip()[23:30]
    seconds_operation = j.rstrip()[27:34]
    if(i+1 <= 10): # n = 50000
        if "build" in BST[i]:
            BST_build[50000].append(seconds_build)
        else:
            BST_operation[50000].append(seconds_operation)
    elif(i+1 >10 and i+1 <= 20):
        if "build" in BST[i]:
            BST_build[100000].append(seconds_build)
        else:
            BST_operation[100000].append(seconds_operation)
    elif(i+1 >20 and i+1 <= 30):
        if "build" in BST[i]:
            BST_build[200000].append(seconds_build)
        else:
            BST_operation[200000].append(seconds_operation)
    else:
        if "build" in BST[i]:
            BST_build[400000].append(seconds_build)
        else:
            BST_operation[400000].append(seconds_operation)
            
print(BST_build)
print(BST_operation)
```

```python
Heap_build = {}
Heap_build[50000] = []
Heap_build[100000] = []
Heap_build[200000] = []
Heap_build[400000] = []

Heap_operation = {}
Heap_operation[50000] = []
Heap_operation[100000] = []
Heap_operation[200000] = []
Heap_operation[400000] = []

for i, j in enumerate(Heap):
    seconds_build = j.rstrip()[24:31]
    seconds_operation = j.rstrip()[28:35]
    if(i+1 <= 10): # n = 50000
        if "build" in Heap[i]:
            Heap_build[50000].append(seconds_build)
        else:
            Heap_operation[50000].append(seconds_operation)
    elif(i+1 >10 and i+1 <= 20): # n = 100,000
        if "build" in Heap[i]:
            Heap_build[100000].append(seconds_build)
        else:
            Heap_operation[100000].append(seconds_operation)
    elif(i+1 >20 and i+1 <= 30):
        if "build" in Heap[i]:
            Heap_build[200000].append(seconds_build)
        else:
            Heap_operation[200000].append(seconds_operation)
    else:
        if "build" in Heap[i]:
            Heap_build[400000].append(seconds_build)
        else:
            Heap_operation[400000].append(seconds_operation)
            
print(Heap_build)
print(Heap_operation)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
plt.title("Build with n = 50,000")
plt.xlabel("# of trials")
plt.ylabel("Seconds")
plt.xticks(x)
plt.plot(x, BST_build[50000], "ro", x, Heap_build[50000], "bs")
plt.axis([0.5, 5.5, 0, 0.018])


avg_b = sum(map(float, BST_build[50000])) / 5
avg_h = sum(map(float, Heap_build[50000])) / 5

plt.text(2.5, 0.0125, "BST build average: " + str(avg_b)[0:5] + " seconds")
plt.text(2.5, 0.002, "Heap build average: "  + str(avg_h)[0:5] + " seconds")

fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/build50000.png", dpi = 100)

plt.show()
```

```python
def avg(list):
    return sum(map(float, list)) / 5

generated = [50000, 100000, 200000, 400000]

# get the data as a dict with generated associated with its average
data_Bb = {}
data_Hb = {}
data_Bo = {}
data_Ho = {}

for num_gen in generated:
    data_Bb[num_gen] = avg(BST_build[num_gen])
    data_Hb[num_gen] = avg(Heap_build[num_gen])
    data_Bo[num_gen] = avg(BST_operation[num_gen])
    data_Ho[num_gen] = avg(Heap_operation[num_gen])
    
    
print(data_Bb)
print(data_Hb)
print(data_Bo)
print(data_Ho)
```

```python
#----------plotting begins--------------

plt.title("Comparison of Build between BST and Heap")
plt.xlabel("# of random number generations to insert into data structure")
plt.ylabel("Seconds")
plt.xticks(generated)

lists_Bb = sorted(data_Bb.items())
x1, y1 = zip(*lists_Bb) # unpack a list of pairs into two tuples
lists_Hb = sorted(data_Hb.items())
x2, y2 = zip(*lists_Hb)


plt.text(200000, 0.09, "BST")
plt.text(200000, 0.01, "Heap")

plt.plot(x1, y1, "r", x2, y2, "b")

fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/build_compare.png", dpi = 100)

plt.show()
```

```python
#----------plotting begins--------------

plt.title("Comparison of Operations between BST and Heap")
plt.xlabel("# of random number generations with operations")
plt.ylabel("Seconds")
plt.xticks(generated)

lists_Bo = sorted(data_Bo.items())
x1, y1 = zip(*lists_Bo) # unpack a list of pairs into two tuples
lists_Ho = sorted(data_Ho.items())
x2, y2 = zip(*lists_Ho)

plt.text(200000, 0.003, "BST")
plt.text(200000, 0.007, "Heap")

plt.plot(x1, y1, "r", x2, y2, "b")


fig = plt.gcf() # gcf(): get current figure
fig.savefig("./Timer/operation_compare.png", dpi = 100)

plt.show()
```
